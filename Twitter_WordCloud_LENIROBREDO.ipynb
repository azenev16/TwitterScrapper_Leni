```python
import os
import matplotlib.pyplot as plt
from wordcloud.wordcloud import WordCloud
```

```python
text = open('lenirobredo.csv',"r")
data = text.read()
print(data)
```

```python
plt.figure(figsize=(15,10))
wordcloud = WordCloud(background_color='white',mode="RGB",width =1000,height =500).generate(data)
plt.title("TOP WORD FOR LENI ROBREDO \n")
plt.imshow(wordcloud)
plt.axis("off")
plt.show()
```
